```python
class CustomSampler(torch.utils.data.sampler.Sampler):
    def __init__(self, mask, data_source):
        self.mask = mask
        self.data_source = data_source
    
    def __iter__(self):
        return iter([i.item() for i in torch.nonzero(self.mask, as_tuple=False)])
    
    def __len__(self):
        return len(self.data_source)
```

```python
def find_landmarks(Y, n, m, j):
    xr = np.linspace(np.min(Y[:,0]), np.max(Y[:, 0]), n)
    yr = np.linspace(np.min(Y[:,1]), np.max(Y[:, 1]), m)
    zr = np.linspace(np.min(Y[:,2]), np.max(Y[:, 2]), j)
    xg, yg, zg = np.meshgrid(xr, yr, zr)
    
    idx = np.zeros(n*m*j, dtype=np.uint)
    
    for i, x, y, z in zip(range(n*m*j), xg.flatten(), yg.flatten(), zg.flatten()):
        idx[i] = np.sum((Y-np.array([x, y, z]))**2, axis=-1).argmin()
        
    return idx
    
def plot_isomap(X, n, m, j):
    fig, axs = plt.subplots(nrows=n*(j//2), ncols=m*(j//2), figsize=(25, 15))
    [axi.set_axis_off() for axi in axs.ravel()]

    for i, ax in enumerate(axs.flatten()):
        img = ax.imshow(np.reshape(X[landmarks[i]], (28, 28)), plt.get_cmap("Greys"))
#         img.set_interpolation("nearest")
    plt.show()
```

# VAE

```python
class Flatten(nn.Module):
    def forward(self, input):
        return input.view(input.size(0), -1)


class Unflatten(nn.Module):
    def __init__(self, channel, height, width):
        super(Unflatten, self).__init__()
        self.channel = channel
        self.height = height
        self.width = width

    def forward(self, input):
        return input.view(input.size(0), self.channel, self.height, self.width)


class ConvVAE(nn.Module):

    def __init__(self, latent_size):
        super(ConvVAE, self).__init__()
        self.beta = 10
        self.latent_size = latent_size

        self.encoder = nn.Sequential(
            nn.Conv2d(1, 64, kernel_size=4, stride=2, padding=1),
            nn.ReLU(),
            nn.Conv2d(64, 128, kernel_size=4, stride=2, padding=1),
            nn.ReLU(),
            Flatten(),
            nn.Linear(6272, 1024),
            nn.ReLU()
        )

        # hidden => mu
        self.fc1 = nn.Linear(1024, self.latent_size)

        # hidden => logvar
        self.fc2 = nn.Linear(1024, self.latent_size)

        self.decoder = nn.Sequential(
            nn.Linear(self.latent_size, 1024),
            nn.ReLU(),
            nn.Linear(1024, 6272),
            nn.ReLU(),
            Unflatten(128, 7, 7),
            nn.ReLU(),
            nn.ConvTranspose2d(128, 64, kernel_size=4, stride=2, padding=1),
            nn.ReLU(),
            nn.ConvTranspose2d(64, 1, kernel_size=4, stride=2, padding=1),
            nn.Sigmoid()
        )

    def encode(self, x):
        h = self.encoder(x)
        mu, logvar = self.fc1(h), self.fc2(h)
        return mu, logvar

    def decode(self, z):
        z = self.decoder(z)
        return z

    
    
      
    def sampling(self, eps, mu, sigma, cond):
        std = torch.exp(0.5 * sigma)
        if not cond:
            eps = torch.randn_like(std) # -----------------------------> pyprob
            return eps.mul(std).add_(mu)
        return eps
    
    def forward(self, x, eps=None, cond=False):
        if cond:
            decoder = self.decoder(eps)
            return decoder, _, _
        
        mu, sigma = self.encode(x)
        z = self.sampling(eps, mu, sigma, cond)
        decoder = self.decoder(z)
        
        return decoder, mu, sigma
    
    
    def loss(self, x_hat, x, mu, sigma):
        # return reconstruction error + KL divergence losses


        BCE = nn.functional.binary_cross_entropy(x_hat, x, reduction='sum')
        KLD = -0.5 * torch.sum(1 + sigma - mu.pow(2) - sigma.exp())

        return BCE + (self.beta*KLD)
```

```python
class VAE(nn.Module):
    def __init__(self, dim=2, cuda=False):
        super(VAE, self).__init__()
        self.dim = dim
        self.beta = 10
        self.device = "cpu"
        if cuda:
            self.device = "cuda"
        self.encoder_ = nn.Sequential(
            nn.Linear(28*28, 200),
            nn.ReLU(),
            nn.Linear(200, 200),
            nn.ReLU(),
            nn.Linear(200, 200),
            nn.ReLU(),
            ).to(self.device)
        
        self.mu_ = nn.Linear(200, self.dim) 
        self.sigma_ = nn.Linear(200, self.dim)
        
        self.decoder_ = nn.Sequential(
            nn.Linear(self.dim, 200),
            nn.ReLU(),
            nn.Linear(200, 200),
            nn.ReLU(),
            nn.Linear(200, 200),
            nn.ReLU(),
            nn.Linear(200, 28*28),
            nn.Sigmoid()
            ).to(self.device)
        
    def encoder(self, x):
        x = self.encoder_(x)
        return self.mu_(x), self.sigma_(x)
        
    
    def decoder(self, x):
        x = self.decoder_(x)
        return x
     
    
    def sampling(self, eps, mu, sigma, cond):
        std = torch.exp(0.5 * sigma)
        if not cond:
            eps = torch.randn_like(std) # -----------------------------> pyprob
            return eps.mul(std).add_(mu)
        return eps
    
    def forward(self, x, eps=None, cond=False):
        if cond:
            decoder = self.decoder(eps)
            return decoder, _, _
        
        mu, sigma = self.encoder(x.view(-1, 28*28))
        z = self.sampling(eps, mu, sigma, cond)
        decoder = self.decoder(z)
        
        return decoder, mu, sigma
    
    def loss(self, x_hat, x, mu, sigma):
        # return reconstruction error + KL divergence losses
        
        
        BCE = nn.functional.binary_cross_entropy(x_hat, x.view(-1, 28*28), reduction='sum')
        KLD = -0.5 * torch.sum(1 + sigma - mu.pow(2) - sigma.exp())
        
        return BCE + (self.beta*KLD)
    
```

```python
vae = VAE(cuda=CFG.cuda)
vae = ConvVAE(2)
optimizer = torch.optim.Adam(vae.parameters())
def train(epoch):
    loss_log = []
    def train_(epoch, loss_log):
        vae.train()
        train_loss = 0
        for batch_idx, (data, label) in enumerate(train_loader):
            if CFG.cuda:
                data = data.to("cuda")
            optimizer.zero_grad()

            recon_batch, mu, log_var = vae(data, label)
            loss = vae.loss(recon_batch, data, mu, log_var)

            loss.backward()
            train_loss += loss.item()
            loss_log.append(loss.item()/len(data))

            optimizer.step()

            if batch_idx % 100 == 0:
                print('\rTrain Epoch: {} [{}/{} ({:.0f}%)]\tLoss: {:.6f}'.format(
                    epoch, batch_idx * len(data), len(train_loader.dataset),
                    100. * batch_idx / len(train_loader), loss.item() / len(data)), flush=False, end="")
        print('\r====> Epoch: {} Average loss: {:.4f}'.format(epoch, train_loss / len(train_loader.dataset)), flush=False, end="")
        
    for epoch in range(epoch):
        train_(epoch, loss_log)
        plt.plot(np.arange(0, len(loss_log)), loss_log , "r", label="Loss", alpha =0.7)
        display.clear_output(wait=True)
        display.display(plt.gcf())
```

```python
train(epoch=15)
```

```python
from pyprob.distributions import Normal
class ProbModel(pyprob.Model):
    def __init__(self, model, loader, dim=2, plot=False):
        super().__init__('VAE')
        self.model = model
        self.dim = dim
        self.loader = loader
        self.plot = plot
    
    def get_batch(self):
        x_ = next(iter(self.loader))
#         x = torch.ones_like(x_[0]) * 0.5
        return x_#x, x_[1]
    
    @torch.no_grad()
    def forward(self):
    
        mu = torch.zeros(100, self.dim)
        sigma = torch.ones(100, self.dim)
        add = pyprob.sample(pyprob_d.Normal(-10, 10), name = "add")
        mul = pyprob.sample(pyprob_d.Normal(-10, 10), name = "mul")

#         mul = pyprob.sample(pyprob_d.Exponential(10), name = "mul")
        eps = add + pyprob.sample(Normal(mu, sigma), name = "eps") * mul
                
        
#         eps_1 = pyprob.sample(Normal(mu, sigma), name = "eps_1")
#         eps_2 = pyprob.sample(Normal(mu, sigma), name = "eps_2")
#         eps = torch.stack((eps_1, eps_2), dim=-1)

        
        data, label = self.get_batch()
        
#         print(label)
        
        sample, _, _ = self.model(data, eps, cond=True)
        
        output=sample.view(-1, 28, 28)

        label[label==3] = 1
        label[label==8] = 0
        
        std =  pyprob.sample(pyprob_d.Exponential(1))
        pyprob.observe(pyprob_d.Normal(output, std), data, name="data")
        if  self.plot:
            fig, axs = plt.subplots(1, 10)
            for i, ax in enumerate(axs.flatten()):
                ax.imshow(output[i],cmap="Greys")
                ax.axis("off")

```
